##### Responda: por que a janela aberta n√£o mostra a imagem colorida?    isso acontece por causa da flag de leitura cv.imread('imagens\lab1_3.jpeg',0) no open cv temos 3 flags de leitura sendo eles

| ----- | --------------------- | ------------------------- |

### Experimento 2 

##### nesse exercicio foi solicitado para que o video ficasse mais r√°pido ou mais lendo, mas na forma que o c√≥digo foi construido tudo vai depender da quantidade atualiza√ß√£o de captura de imagem que esta sendo realizado pelo comando (time.sleep(1/25.0)) sendo assim diminuindo ou aumentando o espa√ßo de captura teremos uma imagem mais fluida ou n√£o 

##### para realizar tudo que foi solicitado iremos transforma a exep√ß√£o do video em uma fun√ß√£o com uma variavel ... e iremos chamo-lo duas vezes, uma com uma velocidade mais rapida e outra mais lenta 

```python

def mostra_video (caminho, tempo):

    cap = cv.VideoCapture(caminho)

    while(cap.isOpened()):
        ret, frame = cap.read()
        
        if ret==True:
            # show the frame
            cv.imshow('frame',frame)

            #wait next frame by 40ms - 25fps
            time.sleep(tempo) 
            
            if cv.waitKey(1) & 0xFF == ord('q'):
                break
        else:
            break
                
    cap.release()
    cv.destroyAllWindows()

mostra_video('videos\lab2.mp4',1/5.0)
mostra_video('videos\lab2.mp4',1/300.0)
```

### Experimento 3 

#### para salvar qualquer imagem pasta acrecentar ao c√≥digo a fun√ß√£o de captura do open cv  -----cv.imwrite("imagens/foto1.png", frame) caso seja necessario salvar varias fotos √© importante tornar o nome da foto uma variavel. 

```python
# ==========================================================
#  ABRIR A C√ÇMERA
# ==========================================================
cap = cv.VideoCapture(0)

if not cap.isOpened():
    print("Cannot open camera")
    exit()

print("Pressione 'x' para salvar a foto.")
print("Pressione 'q' para sair.")

# ==========================================================
#  LOOP DE CAPTURA
# ==========================================================
while True:

    # Captura frame da c√¢mera
    ret, frame = cap.read()

    # Verifica se capturou corretamente
    if not ret:
        print("Erro ao capturar frame.")
        break

    # Mostra imagem na tela
    cv.imshow('Camera', frame)

    # ======================================================
    #  CAPTURA TECLA
    # ======================================================
    tecla = cv.waitKey(1) & 0xFF

    # Se tecla 'x' for pressionada
    if tecla == ord('x'):
        cv.imwrite("imagens/foto1.png", frame)
        print("üì∏ Foto salva como foto1.png")

    # Se tecla 'q' for pressionada
    elif tecla == ord('q'):
        print("Encerrando programa...")
        break


# ==========================================================
# S LIBERAR RECURSOS
# ==========================================================
cap.release()
cv.destroyAllWindows()
```

### Experimento 4 

#### o programa esta mostrando o video invertido pelo comando frame = cv.flip(frame, 0) .... esse comandio pode exibir o video de tres formas 

| Valor | Tipo de Flip                    |
| ----- | ------------------------------- |
| 0     | Vertical (de cabe√ßa para baixo) |
| 1     | Horizontal (espelhado)          |
| -1    | Ambos                           |


```python

# ==========================================================
#  ABRIR C√ÇMERA
# ==========================================================
cap = cv.VideoCapture(0)

if not cap.isOpened():
    print("Erro ao abrir c√¢mera")
    exit()

# ==========================================================
# OBTER DIMENS√ïES
# ==========================================================
width = int(cap.get(cv.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv.CAP_PROP_FRAME_HEIGHT))

# ==========================================================
# OBTER FPS REAL DA C√ÇMERA
# ==========================================================
fps = cap.get(cv.CAP_PROP_FPS)

if fps == 0:
    fps = 30.0  # valor padr√£o seguro

print(f"Gravando com FPS = {fps}")

# ==========================================================
# CONFIGURAR VIDEO WRITER
# ==========================================================
fourcc = cv.VideoWriter_fourcc(*'XVID')
out = cv.VideoWriter('videos/saida.avi', fourcc, fps, (width, height))

# ==========================================================
# LOOP DE CAPTURA
# ==========================================================
while cap.isOpened():

    ret, frame = cap.read()

    if not ret:
        print("Erro ao capturar frame")
        break

    # N√ÉO aplicar flip (imagem normal)
    out.write(frame)

    cv.imshow('frame', frame)

    if cv.waitKey(1) & 0xFF == ord('q'):
        break

# ==========================================================
#  LIBERAR RECURSOS
# ==========================================================
cap.release()
out.release()
cv.destroyAllWindows()
```
