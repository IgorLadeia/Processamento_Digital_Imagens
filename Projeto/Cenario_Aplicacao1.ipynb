# Sistema de Contagem Automática de Comprimidos

Projeto de **Processamento Digital de Imagens (PDI)**.

---

# Tema do Trabalho

**Sistema de Contagem Automática de Comprimidos utilizando Processamento de Imagens**

O sistema utiliza imagens capturadas por câmera ou fotografias para detectar e contar automaticamente comprimidos presentes em uma superfície.

---

# 1. Problema, Contexto, Justificativa e Motivação

Em farmácias, hospitais e até mesmo no uso doméstico, é comum a necessidade de contar comprimidos manualmente para separar doses de medicamentos. Esse processo pode ser demorado e sujeito a erros humanos, especialmente quando há uma grande quantidade de comprimidos.

Em farmácias de manipulação ou hospitais, funcionários frequentemente precisam contar dezenas ou centenas de comprimidos para montar kits de medicação. Além disso, pessoas idosas que utilizam vários medicamentos diariamente podem ter dificuldades para organizar e verificar a quantidade correta.

Dessa forma, surge a necessidade de um sistema simples que utilize **processamento digital de imagens** para identificar e contar automaticamente comprimidos em uma imagem capturada por câmera.

---

# 2. Objetivo do Sistema

O objetivo do sistema é desenvolver uma ferramenta capaz de **identificar e contar automaticamente comprimidos presentes em uma superfície**, utilizando uma câmera ou imagem capturada.

O sistema irá analisar a imagem, detectar os comprimidos presentes e apresentar ao usuário o **número total de comprimidos identificados**.

O resultado final para o usuário será uma **interface simples que mostra a imagem analisada e a quantidade de comprimidos detectados**.

---

# 3. Como se usa o sistema (Interatividade)

O funcionamento do sistema será simples:

1. O usuário posiciona os comprimidos sobre uma superfície plana (mesa ou bandeja).
2. Uma câmera ou smartphone captura a imagem.
3. O sistema processa a imagem utilizando técnicas de processamento digital de imagens.
4. Os comprimidos são detectados e contados automaticamente.
5. O sistema exibe o **número total de comprimidos na tela**.

Opcionalmente, o sistema também pode **destacar os comprimidos detectados na imagem**, facilitando a visualização do resultado.

---

# 4. Benefícios do Sistema ao Usuário

O sistema pode trazer diversos benefícios:

- Redução de erros humanos na contagem de comprimidos.
- Maior rapidez na separação de medicamentos.
- Facilidade de uso para qualquer pessoa.
- Possibilidade de aplicação em **farmácias, hospitais e uso doméstico**.

Além disso, o sistema pode servir como uma ferramenta de apoio para profissionais da saúde, contribuindo para uma **maior segurança na administração de medicamentos**.
