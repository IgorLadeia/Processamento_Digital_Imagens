# Cenário de Aplicação

## Tema do Projeto
Sistema de visão computacional para identificação da posição e orientação (ângulo) de robôs em um campo de futebol, utilizando marcadores geométricos coloridos.

## 1. Problema, Contexto, Justificativa e Motivação

Em competições de robótica, como o futebol de robôs, é fundamental saber com precisão a posição e a orientação de cada robô dentro do campo. Essas informações são essenciais para tomada de decisão, navegação, estratégia de jogo e controle em tempo real.

Entretanto, soluções tradicionais para localização de robôs podem envolver sensores embarcados complexos, como GPS (inviável em ambientes internos), sistemas inerciais ou infraestrutura externa cara. Isso dificulta a implementação em ambientes acadêmicos, especialmente em projetos de graduação.

A partir da análise de cenários educacionais e aplicações em robótica, observa-se a necessidade de uma solução de baixo custo, de fácil implementação e que permita visualizar de forma clara o comportamento dos robôs em campo.

Dessa forma, propõe-se o desenvolvimento de um sistema baseado em processamento digital de imagens, no qual câmeras capturam o campo e algoritmos identificam marcadores geométricos coloridos posicionados sobre os robôs. Esses marcadores permitem determinar tanto a posição quanto o ângulo de orientação de cada robô.

## 2. Objetivo do Sistema

O objetivo do sistema é detectar automaticamente, em tempo real, a posição (coordenadas no campo) e a orientação (ângulo) de um robô de futebol a partir de imagens capturadas por uma câmera.

Como resultado final, o usuário terá uma interface visual que exibe:
- A localização do robô no campo;
- A direção para a qual o robô está orientado;
- Possivelmente o histórico de movimentação (trajetória).

O sistema será projetado para ser compreensível por usuários leigos, especialmente estudantes, permitindo fácil visualização e interpretação dos dados.

## 3. Forma de Uso e Interatividade

O funcionamento do sistema ocorrerá da seguinte forma:

- Uma câmera é posicionada acima ou ao redor do campo de futebol de robôs;
- Cada robô possui marcadores com formas geométricas coloridas (ex: círculos ou combinações de cores);
- O sistema captura imagens em tempo real;
- Técnicas de processamento de imagens são aplicadas para:
  - Detectar as cores dos marcadores;
  - Identificar as formas geométricas;
  - Calcular o centro do robô (posição);
  - Determinar o ângulo de orientação com base na disposição dos marcadores;
- As informações são exibidas em uma interface gráfica.

A interação do usuário consiste principalmente em:
- Inicializar o sistema;
- Ajustar parâmetros (como cores e limiares, se necessário);
- Observar os resultados em tempo real.

## 4. Benefícios para o Usuário

O sistema oferece diversos benefícios:

- Solução de baixo custo, utilizando apenas câmera e software;
- Facilidade de implementação em ambientes acadêmicos;
- Auxílio no ensino de conceitos de robótica e processamento de imagens;
- Visualização clara e intuitiva da posição e orientação dos robôs;
- Possibilidade de uso em estratégias de controle e automação em tempo real;
- Base para desenvolvimento de projetos mais avançados em visão computacional e robótica.

Esse sistema contribui tanto para aplicações educacionais quanto para experimentação em robótica, tornando o processo mais acessível e didático.
